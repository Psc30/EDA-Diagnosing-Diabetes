12. To figure out why the `Outcome` column is of type `object` (string) instead of type `int64`, print out the unique values in the `Outcome` column.

````python
# print unique values of Outcome column
diabetes_data.Outcome.unique()
````

13. How might you resolve this issue?

**Your response to question 13**: A possible next step would be to replace instances of 'O' with 0 and convert the `Outcome` column to type int64.

````python
# Replace instances of 'O' with 0 in the 'Outcome' column and then convert it into type int64.
diabetes_data['Outcome'] = diabetes_data['Outcome'].replace('O', 0).astype('int64')
diabetes_data.info()
diabetes_data.Outcome.unique()
````

    - Use `.value_counts()` to more fully explore the values in each column.
    - Instead of changing the `0` values in the five columns to `NaN`, try replacing the values with the median or mean of each column.

````python
# Use .value_counts() to explore the values in each column.
for column in diabetes_data.columns:
    print(f'Value counts for {column}:')
    print(diabetes_data[column].value_counts())
    print('\n')
````

````python
# Instead of changing the 0 values in the five columns to NaN, try replacing the values with the mean of each column.
columns_to_replace = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
# Calculate the mean and replace 0 values.
for column in columns_to_replace:
    mean_value = diabetes_data[column].mean()
    diabetes_data[column] = diabetes_data[column].replace(0, mean_value)
diabetes_data.describe()
````
